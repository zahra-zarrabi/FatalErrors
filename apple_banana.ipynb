```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['legend.fontsize'] = 15
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 15
```

```python
n=1000
std=1.5

bananas=pd.DataFrame({'length':np.random.normal(8,std,n),'width':np.random.normal(4,std,n),'class':np.zeros(n,dtype='int')})

apple=pd.DataFrame({'length':np.random.normal(6,std,n),'width':np.random.normal(6,std,n),'class':np.ones(n,dtype='int')})

fruits = pd.concat([bananas,apple])
```

```python
plt.scatter(x=bananas['length'],y=bananas['width'],label='bananas',c='orange')
plt.scatter(x=apple['length'],y=apple['width'],label='apple',c='red')
plt.legend()
plt.xlabel('Length')
plt.ylabel('Width')
```

```python
labels = ['bananas', 'apple', 'bananas', 'apple']
x = [2.5, 3, 4, 4.5]
height = [fruits.loc[fruits['class'] == 0, 'length'].mean(),
          fruits.loc[fruits['class'] == 1, 'length'].mean(),
          fruits.loc[fruits['class'] == 0, 'width'].mean(),
          fruits.loc[fruits['class'] == 1, 'width'].mean()]
yerr = [2*fruits.loc[fruits['class'] == 0, 'length'].std(),
        2*fruits.loc[fruits['class'] == 1, 'length'].std(),
        2*fruits.loc[fruits['class'] == 0, 'width'].std(),
        2*fruits.loc[fruits['class'] == 1, 'width'].std()]
color = ['orange','red',  'tab:orange', 'tab:red']

plt.bar(x=x,
        height=height,
        color=color,
        width=0.4,
        error_kw={'lw': 4},
        yerr=yerr
       )
plt.xticks(ticks=x, labels=labels)
plt.title('Fruit length VS width')

plt.grid()
```

```python
fea = 'length'
plt.boxplot(x=fea, data=fruits.loc[fruits['class'] == 0], patch_artist=True, boxprops={'facecolor': 'orange'}, positions=[0], labels=['bananas'], widths=[0.15])
plt.boxplot(x=fea, data=fruits.loc[fruits['class'] == 1], patch_artist=True, boxprops={'facecolor': 'red'}, positions=[0.3], labels=['apples'], widths=[0.15])
plt.ylabel('length')
plt.xlim(-0.2, 0.5)
plt.title('Length Distribution ')
```

```python
fea = 'width'

plt.hist(x=fea, data=fruits.loc[fruits['class'] == 0],bins=30, density=True, color='tab:orange', label='bananas',)
plt.hist(x=fea, data=fruits.loc[fruits['class'] == 1], bins=30, density=True, color='tab:red', label='apples', alpha=0.5)
plt.legend(frameon=False)
plt.title('Width Distribution ')
plt.xlabel('Width')
```

```python
fig, (ax, cax) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]})
_, _, _, mappable = ax.hist2d( x='length', y='width', bins=50, data=fruits, cmap=plt.cm.hot)
ax.set_xlabel('length')
ax.set_ylabel('width')

plt.colorbar(mappable=mappable, cax=cax, ax=ax)
```

```python
fea = 'L/W'

fruits[fea] = fruits['length']/fruits['width']

plt.hist(x=fea, data=fruits.loc[fruits['class'] == 0],bins=[-2, -1, 0, 1, 2, 3, 4, 5, 6, 7, 8], density=True, color='orange', label='bananas')
plt.hist(x=fea, data=fruits.loc[fruits['class'] == 1],bins=[-2, -1, 0, 1, 2, 3, 4, 5, 6, 7, 8], density=True, color='red', label='apples',alpha=0.5)
plt.legend()
plt.title('L/W Distribution ')
plt.xlabel('L/W ratio')
```

```python
fea='area'
fruits['area'] = fruits['length']*fruits['width']
plt.hist(x=fea,data=fruits.loc[fruits['class'] == 0], histtype='step', lw=4,bins=30, density=True, color='orange', label='bananas')
plt.hist(x=fea, data=fruits.loc[fruits['class'] == 1], histtype='step', lw=4,bins=30, density=True, color='red', label='apples')
plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')
```

```python
fea = 'circumference'
fruits[fea] = 2*(fruits['length']+fruits['width'])
plt.hist(x=fea, data=fruits.loc[fruits['class'] == 0], histtype='step', lw=4,bins=30, density=True, color='orange', label='bananas')
plt.hist(x=fea, data=fruits.loc[fruits['class'] == 1], histtype='step', lw=4,bins=30, density=True, color='red', label='apple')
plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('Circumference')
```
